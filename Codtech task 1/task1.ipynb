```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("bigdata_sample.csv")

# 1. Trip Distance Distribution
df['trip_distance'].plot(kind='hist', bins=40, figsize=(8,5), color='skyblue', title="Trip Distance Distribution")
plt.xlabel("Distance (miles)")
plt.ylabel("Number of Trips")
plt.show()

# 2. Passenger Count
df['passenger_count'].value_counts().sort_index().plot(kind='bar', figsize=(8,5), color='orange', title="Passenger Count per Trip")
plt.xlabel("Passenger Count")
plt.ylabel("Number of Trips")
plt.show()

# 3. Payment Type Distribution
df['payment_type'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6), title="Payment Type Distribution")
plt.ylabel("")
plt.show()

# 4. Fare vs Tip Scatter
df.plot(kind="scatter", x="fare_amount", y="tip_amount", alpha=0.3, color="green", figsize=(8,5), title="Fare vs Tip Amount")
plt.show()

# 5. Revenue Insights (Text only)
print("Average Fare:", df['fare_amount'].mean())
print("Average Tip:", df['tip_amount'].mean())
print("Average Total:", df['total_amount'].mean())


```
